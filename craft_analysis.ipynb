```python
df4 = pd.read_csv('D:\Feynn labs - Internship\data\Year-wise-production-of-handicrafts-Rs-in-lacs.png.csv').drop('Y=715.80 + 73.92X (Trend value of production)', axis=1)
df4.columns = df4.columns.str.strip()
df4.head()
```

```python
df5 = pd.read_csv('D:\Feynn labs - Internship\data\Year-wise-sales-of-handicrafts-Rs-in-lacs.png.csv').drop('Y = 750.80 + 74.05X (Trend value of Sales)', axis=1)
df5.columns = df5.columns.str.strip()
df5.head()
```

```python
df4.info() # production year-wise
```

```python
# set index for year column
df4.set_index('Year', inplace=True)
df4.index.name = 'year'
# Yearly average production calculation
df4['yearly_average_production'] = df4.mean(axis=1)
df4
```

```python
# SMA over a period of 10 and 20 years 
df4['SMA_10'] = df4.yearly_average_production.rolling(10, min_periods=1).mean()
df4['SMA_20'] = df4.yearly_average_production.rolling(20, min_periods=1).mean()
```

```python
# Grean = Avg Prod, RED = 10 yrs, ORANG colors for the line plot
colors = ['green', 'red', 'orange']
# Line plot 
df4.plot(color=colors, linewidth=3, figsize=(12,6))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['Average production value', '10-years SMA', '20-years SMA'], fontsize=14)
plt.title('The yearly average production value (Rs. Lakhs)', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Production Value (Lacs of Rs.)', fontsize=16)
```
